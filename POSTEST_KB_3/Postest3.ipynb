# 1. Mengimport Library pandas, matplotlib dan seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2. Menampilkan dataset yang dimiliki

```python
Netflix = pd.read_csv("netflix daily top 10.csv")
Netflix.info()
```

# 3. Grafik batang berdasarkan type
**Jenis Acara yang paling banyak dilihat atau ditonton oleh banyak orang adalah TV Show**

```python
x = list(Netflix["Type"])
y = list(Netflix["Viewership Score"])

plt.figure(figsize = (10,5))
plt.bar(x, y, label = "Type", color = ["lightblue"])
plt.xlabel("Type Movies")
plt.ylabel("Total type Movies")
plt.title("Total keseluruhan type Movies")
plt.legend()
plt.show()
```

# 4. Pie chart berdasarkan type
**Grafik di bawah ini menunjukkan bahwasannya TV Show sangatlah banyak diminati yakni sekitar 62.62% dibandingkan Jenis Acara lainnya, berbanding terbalik dengan Jeni Acara Concert/Performance yang memiliki sedikit peminat yakni sekitar 0.03%**

```python
data = Netflix.Type.value_counts()
plt.pie(data, labels=data.index, explode = (0.1, 0.1, 0.1, 0), startangle=90, autopct="%1.1f%%")
plt.title('Jumlah Data Penonton terhadap jenis acara')
plt.show()
```

# 5. Membuat Grafik Boxplot berdasarkan type dan viewership score
**Disini ditampilkan grafik type b viewership score yang dimana data miningnya tidak simetris dikarenakan median di dalam kotak tidak berada di tengah dan garis atas lebih panjang**

```python
import seaborn as sns
sns.boxplot(x = Netflix['Type'], y = Netflix['Viewership Score'])
plt.xlabel("Movie Type")
plt.ylabel("rata-rata viewership score")
plt.title("Nilai viewership score masing-masing type movie ")
plt.show()
```

# 6. Membuat Heatmap berdasarkan seluruh data didalam dataset
**Disini ditampilkan korelasi antara data days in top 10 dan viewership score memiliki angka yang cukup tinggi**

```python
cor_matrix = Netflix.corr()
sns.heatmap(cor_matrix, cmap='coolwarm', annot=True, center=0)
plt.title("Korelasi antara data")
plt.show()
```
