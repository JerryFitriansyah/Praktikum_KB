## 3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin.

```python
Netflix.info()
record, attribute = Netflix.shape
jumlah_angka = Netflix.select_dtypes(["int64", "float64"]).any().sum()
nama_attribute = Netflix.columns.values

hasil = {
    "Jumlah Record Keseluruhan" : record,
    "\nJumlah Attribute Keseluruhan" : attribute,
    "\nNama Attribute" : list(nama_attribute),
    "\nJumlah Attribute (Tipe Angka)" : jumlah_angka,
}

for x in hasil:
    print(x, ":", hasil[x])
```

## 5.Tampilkan jumlah nilai dari semua attribute yang bernilai object (contoh : Jenis Kelamin, Agama).

```python
kolom_object = Netflix.select_dtypes(include=['object'])
count_results = {}
for kolom in kolom_object:
    count_results[kolom] = Netflix[kolom].value_counts()
for kolom, hasil in count_results.items():
    print("=======================================")
    print(f"Jumlah nilai dalam {kolom}:")
    print("=======================================")
    print(hasil, "\n")
```
